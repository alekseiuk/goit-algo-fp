```python
import random
import pandas as pd
import matplotlib.pyplot as plt
```

```python
N = 1_000_000
```

```python
counts = {s: 0 for s in range(2, 13)}
```

```python
for _ in range(N):
    d1 = random.randint(1, 6)
    d2 = random.randint(1, 6)
    counts[d1 + d2] += 1
```

```python
probabilities = {s: counts[s] / N * 100 for s in counts}
```

```python
df = pd.DataFrame({
    "Сума": list(counts.keys()),
    "Імовірність": list(probabilities.values())
})

print(df.to_string(index=False))
```
