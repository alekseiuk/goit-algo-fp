```python
from ColabTurtlePlus.Turtle import *
```

```python
def draw_pythagoras_tree(length, level):
    """
    У ColabTurtlePlus ми використовуємо глобальні методи,
    або методи об'єкта, тут для простоти використовуємо стиль "turtle."
    """
    if level == 0:
        return

    forward(length)

    left(45)

    draw_pythagoras_tree(length * 0.7, level - 1)

    right(90)

    draw_pythagoras_tree(length * 0.7, level - 1)

    left(45)

    backward(length)
```

```python
clearscreen()
setup(500, 500)       # Розмір полотна
bgcolor("white")
speed(13)             # 13 - це максимальна швидкість у цій бібліотеці

# Налаштування "черепашки"
color("green")
width(2)

# Позиціонування (знизу екрану)
penup()
goto(0, -200)
pendown()
left(90)
```

```python
recursion_level = 8
print(f"Малюємо дерево рівня {recursion_level}...")

draw_pythagoras_tree(100, recursion_level)
```
